image.png

Na biblioteca Pydeequ, temos três tipos principais de analyzers:
1. Do dataset todo
2. De colunas específicas
3. De comparação

Analyzers de dataset são aqueles que trazem informações sobre todo o dataset. O mais simples deles é o Size(), que retorna a quantidade de linhas do dataset

Analysers de colunas específicas trazem informações daquelas colunas. Um exemplo é o Completeness(), que retorna a razão de linhas não nulas.

Annalysers de comparação comparam informações de duas ou mais colunas do dataset. Um exemplo é o Correlation(), que retorna a correlação de pearson dada duas colunas.
